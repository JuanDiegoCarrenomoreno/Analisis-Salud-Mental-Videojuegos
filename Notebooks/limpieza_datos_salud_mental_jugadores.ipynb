# Limpieza de Datos Sobre Videojuegos y Salud Mental

Este notebook se enfoca en la limpieza y preprocesamiento de un conjunto de datos relacionado con hábitos de juego, indicadores de salud mental, calidad del sueño y rendimiento académico en jugadores.

## Objetivos
- Limpiar datos inconsistentes
- Eliminar variables innecesarias
- Estandarizar valores categóricos
- Validar coherencia lógica entre métricas
- Preparar el dataset para visualización y análisis

## Carga del Dataset

El conjunto de datos es importado e inspeccionado para comprender su estructura, variables y posibles inconsistencias antes del proceso de limpieza. Una vez realizada la inspección, se modifican y se traducen los títulos para una mejor comprensión de parte del lector.

```python
import pandas as pd
import numpy as np

df = pd.read_csv("gaming_mental_health.csv")

titulos = ["ID", "Edad", "Genero Jugador", "Horas Jugadas Diarias", "Genero Juego", "Juego Principal", "Plataforma Jugada", "Horas De Sueno", "Calidad De Sueno", "Interrupcion De Sueno", "Rendimiento Academico", "Notas", "Puntos Productividad Laboral", "Estado De Animo", "Frecuencia A Cambios De Humor", "Sindromes De Abstinencia", "Perdida De Otros Intereses", "Continuo A Pesar De Los Problemas", "Fatiga Visual", "Dolor De Espalda O Cuello", "Cambios En Peso Kg", "Horas De Ejercicio Semanalmente", "Puntos Aislamiento Social", "Horas De Encuentros Sociales Semanalmente", "Gastos En Juegos Mensualmente En Dolares", "Anos Jugando", "Riesgo De Adiccion A Los Juegos"]
df.columns = titulos
df.head()
```

## Manejo de Valores Nulos

Se eliminan registros incompletos o con valores nulos para garantizar consistencia durante el análisis de datos.

```python
df.dropna(subset = ['Notas','Puntos Productividad Laboral'], inplace=True)
```

## Selección de Variables

Se eliminan columnas poco relevantes o innecesarias con el fin de simplificar el dataset y enfocar el análisis en las variables principales.

```python
elim_cols = ["Calidad De Sueno", "Interrupcion De Sueno","Puntos Productividad Laboral","Cambios En Peso Kg","Puntos Aislamiento Social","Gastos En Juegos Mensualmente En Dolares"]
df.drop(elim_cols, axis=1, inplace=True)
df
```

## Traducción de Variables y Valores

Se realizó la traducción de nombres de columnas y valores categóricos originalmente en inglés al español con el objetivo de mejorar la comprensión, legibilidad y consistencia general del dataset durante el proceso de análisis.

```python
df["Genero Jugador"] = df["Genero Jugador"].replace({"Male":"Masculino","Female":"Femenino","Other":"Otro"})
df["Plataforma Jugada"] = df["Plataforma Jugada"].replace({"Multi-platform":"Multi-platforma","Console":"Consola","Mobile":"Celular"})
df["Genero Juego"] = df["Genero Juego"].replace({"Strategy":"Estrategia","Mobile Games":"Juegos De Celular"})
df
```

## Validación de Calidad del Sueño

Las categorías relacionadas con la calidad del sueño fueron reajustadas según las horas de sueño registradas para mejorar la coherencia lógica de los datos.

```python

bins = [0, 4, 5.5, 6.5, 8, 10]
labels = ["Muy Baja", "Baja", "Intermedia", "Buena", "Muy buena"]

df["Calidad Sueno"] = pd.cut(
    df["Horas De Sueno"],
    bins=bins,
    labels=labels,
    right=False
)

df[["Horas De Sueno", "Calidad Sueno"]].head(10)
```

```python
df["Calidad Sueno"] = df["Calidad Sueno"].astype(str)
df.info()
```

## Validación del Rendimiento Académico

Las categorías de rendimiento académico fueron recalculadas a partir de las notas obtenidas para reducir inconsistencias entre variables numéricas y categóricas.

```python
df[["Rendimiento Academico", "Notas"]].head(10)
```

```python
bins = [0, 2.5, 3, 4, 4.5, 5.1]
labels = ["Insuficiente", "Bajo", "Medio", "Alto", "Superior"]

df["Rendimiento Academico"] = pd.cut(
    df["Notas"],
    bins=bins,
    labels=labels,
    right=False
)

df["Rendimiento Academico"].value_counts()
```

```python
df[["Rendimiento Academico", "Notas"]].head(10)
```

### Observación

Durante el proceso de limpieza se identificaron inconsistencias entre algunas variables categóricas y sus respectivos valores numéricos, por lo que fue necesario realizar ajustes para mejorar la calidad del dataset.

## Dataset Final

El conjunto de datos se encuentra limpio, estandarizado y preparado para procesos posteriores de análisis exploratorio y visualización.

```python
df.to_csv("gaming_mental_health_cleaning.csv", index=False)
```
